# Matplotlib 绘图

```python
import matplotlib
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection

import numpy as np

import matplotlib.pyplot as plt

# 绘制一个多边形
polygon = Polygon(
    [(1, 1), (2, 2), (1, 3)],
    closed=True,
    color='blue',
    fill=None
)
# 创建一个图形
fig, ax = plt.subplots()
# 添加多边形到图形
ax.add_patch(polygon)
# 设置坐标轴范围
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
# 显示图形
plt.show()
```
